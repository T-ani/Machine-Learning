```python
df =  pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
```

```python
X_N = df['area'].values.reshape(-1,1)
Y_N = df['price'].values
reg.fit(X_N,Y_N)
```

```python
reg.coef_
```

```python
reg.intercept_
```
